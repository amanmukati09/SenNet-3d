# Sample Images Path
## kidney 3 dense doesn't have image it has label only

```python
def showImage(pth,img_name):
    
    image_1 = imgo.imread(f"{path_train}/{pth}/images/{img_name}.tif")
    label_1 = imgo.imread(f"{path_train}/{pth}/labels/{img_name}.tif")
    
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(image_1, cmap='gray')
    plt.axis('off')
    plt.title("Image")

    plt.subplot(1,2,2)
    plt.imshow(label_1, cmap='gray')
    plt.axis('off')
    plt.title("Label")
    
    plt.show()
```

# Kidney 1 Dense 0000.tif

```python
showImage("kidney_1_dense","0000")
```

# Kidney 1 VOI 0000.tif
